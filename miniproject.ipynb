```python
import pandas as pd 
```

```python
df=pd.read_csv("netflix.csv")
```

```python
clean=df.dropna()
```

```python
clean.isnull().sum()
```

```python
clean.duplicated()
```

```python
drop=clean.drop_duplicates()
drop
```

```python
drop.rename(columns={"date_added":"dateTime"},inplace=True)
drop
```

```python
drop["dateTime"]=drop["dateTime"].astype(str).str.strip()

drop["dateTime"]=pd.to_datetime(drop["dateTime"],errors='coerce')

```

```python
drop["year_added"]=drop["dateTime"].dt.year
```

```python
drop["month_added"]=drop["dateTime"].dt.month
```

```python
drop
```

```python
drop.groupby("type").count()
```

```python
drop.groupby("country").count()
```

```python
list_of_countries = drop["country"].value_counts().head(10)
print(list_of_countries)
```

```python
common_genre=drop["listed_in"].value_counts().head(10)
common_genre
```
